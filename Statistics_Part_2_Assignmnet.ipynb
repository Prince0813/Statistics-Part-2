# Theory Question

 1. What is hypothesis testing in statistics?
- Hypothesis testing is a statistical method used to determine whether there is enough evidence in a sample data to infer that a certain condition is true for the entire population.

 2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
- The null hypothesis (H₀) states that there is no effect or no difference, while the alternative hypothesis (H₁) suggests that there is a significant effect or difference.

 3. What is the significance level in hypothesis testing, and why is it important?
- The significance level (α) is the probability of rejecting the null hypothesis when it is actually true. It is important because it determines the threshold for statistical significance, commonly set at 0.05 or 0.01.

 4. What does a P-value represent in hypothesis testing?
- The P-value represents the probability of obtaining test results at least as extreme as the observed data, assuming that the null hypothesis is true.

 5. How do you interpret the P-value in hypothesis testing?
- If the P-value is less than the significance level (α), we reject the null hypothesis. If it is greater than α, we fail to reject the null hypothesis.

 6. What are Type 1 and Type 2 errors in hypothesis testing?
- Type 1 error (False Positive): Rejecting the null hypothesis when it is actually true.
- Type 2 error (False Negative): Failing to reject the null hypothesis when it is actually false.

 7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
- One-tailed test: Tests for an effect in only one direction.
- Two-tailed test: Tests for an effect in both directions.

 8. What is the Z-test, and when is it used in hypothesis testing?
- A Z-test is used when the sample size is large (n > 30) and the population variance is known. It tests whether a sample mean significantly differs from a population mean.

 9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
- The Z-score is calculated as:where X is the sample mean, μ is the population mean, and σ is the standard deviation. It represents how many standard deviations the sample mean is from the population mean.

 10. What is the T-distribution, and when should it be used instead of the normal distribution?
- The T-distribution is used when the sample size is small (n < 30) and the - - - population standard deviation is unknown.

 11. What is the difference between a Z-test and a T-test?
- Z-test: Used for large samples with known population variance.
- T-test: Used for small samples with unknown population variance.

 12. What is the T-test, and how is it used in hypothesis testing?
- A T-test is used to compare the means of two groups and determine if they are significantly different from each other.

 13. What is the relationship between Z-test and T-test in hypothesis testing?
- Both tests assess the difference between means, but the T-test is more appropriate for small samples, while the Z-test is suitable for large samples with known variance.

 14. What is a confidence interval, and how is it used to interpret statistical results?
- A confidence interval is a range of values within which the true population parameter is expected to fall with a certain level of confidence (e.g., 95%).

 15. What is the margin of error, and how does it affect the confidence interval?
- The margin of error is the range above and below the sample statistic that defines the confidence interval. A smaller margin of error leads to a more precise estimate.

 16. How is Bayes' Theorem used in statistics, and what is its significance?
- Bayes' Theorem calculates the probability of an event based on prior knowledge of related conditions. It is used in Bayesian statistics for updating beliefs based on evidence.

 17. What is the Chi-square distribution, and when is it used?
- The Chi-square distribution is used in hypothesis testing for categorical data, such as independence tests and goodness-of-fit tests.

 18. What is the Chi-square goodness of fit test, and how is it applied?
- The Chi-square goodness-of-fit test checks if a sample distribution matches an expected distribution.

 19. What is the F-distribution, and when is it used in hypothesis testing?
- The F-distribution is used in variance analysis, such as ANOVA and regression analysis, to compare multiple group variances.

 20. What is an ANOVA test, and what are its assumptions?
- ANOVA (Analysis of Variance) tests whether the means of three or more groups are significantly different. Assumptions include normality, independence, and equal variance.

 21. What are the different types of ANOVA tests?
- One-way ANOVA: Compares means of one independent variable.
- Two-way ANOVA: Compares means of two independent variables.
- Repeated Measures ANOVA: Used when the same subjects are tested multiple times.


 22. What is the F-test, and how does it relate to hypothesis testing?
- The F-test compares variances of different groups to determine if they are significantly different. It is commonly used in ANOVA and regression analysis.

```python
#1 Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the resultsimport numpy as np
from scipy.stats import norm

def one_sample_z_test(sample, pop_mean, pop_std):
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    print(f"Z-score: {z_score}")
    print(f"P-value: {p_value}")

    if p_value < 0.05:
        print("Reject the null hypothesis.")
    else:
        print("Fail to reject the null hypothesis.")

# Simulate data
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)
one_sample_z_test(sample_data, pop_mean=50, pop_std=10)
```

```python
#2 Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np # Importing numpy
from scipy.stats import norm # Importing norm


def plot_z_test_decision_region(z_score):
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    plt.figure(figsize=(8, 5))
    # Pass x and y as a dictionary to the 'data' argument
    sns.lineplot(x=x, y=y, label="Standard Normal Distribution")

    plt.fill_between(x, y, where=(x < -1.96) | (x > 1.96), color='red', alpha=0.5, label="Rejection Region")
    plt.axvline(z_score, color='blue', linestyle='--', label=f"Z-score: {z_score}")

    plt.legend()
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.title("Two-Tailed Z-test Decision Region")
    plt.show()

# Perform Z-test
sample_data = np.random.normal(loc=52, scale=10, size=30)
sample_mean = np.mean(sample_data)
z_score = (sample_mean - 50) / (10 / np.sqrt(30))

plot_z_test_decision_region(z_score)
```

```python
#3 Implement a one-sample Z-test using Python to compare the sample mean with the population mean

def visualize_errors():
    import numpy as np
    import seaborn as sns
    import matplotlib.pyplot as plt
    from scipy.stats import norm

    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    plt.figure(figsize=(8, 5))
    # Pass x and y as keyword arguments (or as a dictionary to 'data')
    sns.lineplot(x=x, y=y, label="Null Distribution (H0)")
    plt.fill_between(x, y, where=(x > 1.96) | (x < -1.96), color='red', alpha=0.5, label="Type 1 Error (α)")

    alternative_x = np.linspace(-4, 4, 1000)
    alternative_y = norm.pdf(alternative_x, 1, 1)
    # Pass alternative_x and alternative_y as keyword arguments (or as a dictionary to 'data')
    sns.lineplot(x=alternative_x, y=alternative_y, label="Alternative Distribution (H1)")
    plt.fill_between(alternative_x, alternative_y, where=(alternative_x < 1.96) & (alternative_x > -1.96), color='blue', alpha=0.5, label="Type 2 Error (β)")

    plt.legend()
    plt.xlabel("Test Statistic")
    plt.ylabel("Probability Density")
    plt.title("Type 1 and Type 2 Errors")
    plt.show()

visualize_errors()
```

```python
#4 Perform a two-tailed Z-test using Python and visualize the decision region on a plot

from scipy.stats import ttest_ind

def independent_t_test(sample1, sample2):
    t_stat, p_value = ttest_ind(sample1, sample2, equal_var=False)

    print(f"T-statistic: {t_stat}")
    print(f"P-value: {p_value}")

    if p_value < 0.05:
        print("Reject the null hypothesis.")
    else:
        print("Fail to reject the null hypothesis.")

# Simulate data
np.random.seed(42)
group1 = np.random.normal(50, 10, 30)
group2 = np.random.normal(55, 10, 30)

independent_t_test(group1, group2)
```

```python
#5 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

from scipy.stats import ttest_rel

def paired_t_test(sample1, sample2):
    t_stat, p_value = ttest_rel(sample1, sample2)

    print(f"T-statistic: {t_stat}")
    print(f"P-value: {p_value}")

    if p_value < 0.05:
        print("Reject the null hypothesis.")
    else:
        print("Fail to reject the null hypothesis.")

    plt.figure(figsize=(6, 4))
    plt.scatter(sample1, sample2)
    plt.xlabel("Before")
    plt.ylabel("After")
    plt.title("Paired T-test Comparison")
    plt.show()

# Simulate data
np.random.seed(42)
before = np.random.normal(50, 10, 30)
after = before + np.random.normal(2, 5, 30)

paired_t_test(before, after)
```

```python
#6 Write a Python program to perform an independent T-test and interpret the results

from scipy.stats import ttest_1samp

def compare_z_t_tests(sample, pop_mean, pop_std):
    # Z-test
    z_score = (np.mean(sample) - pop_mean) / (pop_std / np.sqrt(len(sample)))
    z_p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # T-test
    t_stat, t_p_value = ttest_1samp(sample, pop_mean)

    print(f"Z-test P-value: {z_p_value}")
    print(f"T-test P-value: {t_p_value}")

# Simulate data
sample_data = np.random.normal(50, 10, 30)
compare_z_t_tests(sample_data, pop_mean=50, pop_std=10)
```

```python
#7 Perform a paired sample T-test using Python and visualize the comparison results

import scipy.stats as stats

def confidence_interval(sample, confidence=0.95):
    mean = np.mean(sample)
    std_error = stats.sem(sample)
    margin_of_error = std_error * stats.t.ppf((1 + confidence) / 2, len(sample) - 1)

    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    print(f"Confidence Interval: ({lower_bound}, {upper_bound})")
    return lower_bound, upper_bound

# Simulate data
sample_data = np.random.normal(50, 10, 30)
confidence_interval(sample_data)
```

```python
#8 Simulate data and perform both Z-test and T-test, then compare the results using Python

import numpy as np
from scipy import stats

# Simulate data
np.random.seed(0)
data_1 = np.random.normal(50, 10, 100)  # sample 1
data_2 = np.random.normal(55, 10, 100)  # sample 2

# Z-test (assuming population std deviation is known)
z_stat, p_value_z = stats.ttest_1samp(data_1, 50)

# T-test (for independent samples)
t_stat, p_value_t = stats.ttest_ind(data_1, data_2)

print("Z-test statistic:", z_stat, "P-value:", p_value_z)
print("T-test statistic:", t_stat, "P-value:", p_value_t)
```

```python
#9 Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    std_err = stats.sem(data)
    margin_of_error = std_err * stats.t.ppf((1 + confidence) / 2., len(data) - 1)
    return mean - margin_of_error, mean + margin_of_error

data = np.random.normal(50, 10, 100)
ci = confidence_interval(data)
print("Confidence interval:", ci)
```

```python
#11 Implement a Bayesian inference method using Bayes'Theorem in Python and explain the process

def bayes_theorem(prior, likelihood, evidence):
    return (likelihood * prior) / evidence

prior = 0.3  # Probability of hypothesis being true
likelihood = 0.7  # Probability of observing the data given the hypothesis
evidence = 0.5  # Probability of the data

posterior = bayes_theorem(prior, likelihood, evidence)
print("Posterior probability:", posterior)
```

```python
#19 Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

import scipy.stats as stats
import numpy as np

def check_anova_assumptions(data):
    normality_p_values = [stats.shapiro(group)[1] for group in data]
    equal_variance = stats.levene(*data)[1]
    print("Normality p-values:", normality_p_values)
    print("Equal variance p-value:", equal_variance)

group1 = np.random.normal(50, 10, 100)
group2 = np.random.normal(55, 10, 100)
group3 = np.random.normal(60, 10, 100)

check_anova_assumptions([group1, group2, group3])
```

```python
#20 Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results

import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns
import numpy as np # Importing NumPy for random data generation
import pandas as pd # Importing pandas and aliasing it as 'pd'


# Simulate two-way ANOVA data
data = {
    'factor1': np.random.choice(['A', 'B'], 200),
    'factor2': np.random.choice(['X', 'Y'], 200),
    'value': np.random.normal(50, 10, 200)
}

df = pd.DataFrame(data) # Creating DataFrame using pd.DataFrame
model = ols('value ~ C(factor1) + C(factor2) + C(factor1):C(factor2)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print(anova_table)

# Plot interaction
sns.boxplot(x="factor1", y="value", hue="factor2", data=df)
plt.show()
```

```python
#24 Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

import numpy as np
import scipy.stats as stats

data = np.random.normal(50, 10, 100)
variance = np.var(data)
chi2_stat = (len(data) - 1) * variance / 100

p_value = stats.chi2.sf(chi2_stat, len(data) - 1)
print("Chi-square statistic:", chi2_stat, "P-value:", p_value)
```
